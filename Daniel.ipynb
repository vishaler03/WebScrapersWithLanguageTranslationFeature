```python
def price_page_handler():
    
    global driver
    
    driver.implicitly_wait(10)
    
    price=driver.find_element_by_xpath('/html/body/div[1]/div/header/div[2]/nav/ul/li[2]/div/span')
    
    price.click()
    
    sleep(2)
    
    try:
    
        x=driver.find_element_by_xpath('//*[@id="product-form-sale-price--fi"]')
        
        x.click()
        
        for i in range(10):
            
            x.send_keys(Keys.BACKSPACE)
            
        x.send_keys('0')
            
    except: pass
    
    try:
        
        y=driver.find_element_by_xpath('//*[@id="product-form-sale-price--dk"]')
        
        y.click()
        
        for i in range(10):
            
            y.send_keys(Keys.BACKSPACE)
            
        y.send_keys('0')
        
    except: pass
    
    try:
        
        vat=driver.find_elements_by_xpath('//div[@class="mc-multi-radio-buttons__option mc-multi-radio-buttons__option--selected"]')
        
        for i in vat:
            
            i.click()
            
            sleep(1)
            
    except: pass   
```

```python
def total_countries():
    
    global driver
    
    driver.implicitly_wait(20)
    
#     wd.support.wait.WebDriverWait.until(10)
    
    tab=driver.find_element_by_xpath('//div[@class="product-comparison__tabs-wrapper"]')
    
    li=tab.find_elements_by_tag_name('li')
    
    res=[]
    
    for i in li:
        
        res.append(i.text)
        
    return len(res)
    
    print(len(li))
    
```

```python
def filtering():
    
    global driver
    
    driver.implicitly_wait(20)
    
#     wd.support.wait.WebDriverWait.until(20)
    
    try:
        
        products=driver.find_element_by_xpath('//*[@id="sidebar"]/div[3]/a[2]')
    
        products.click()
    
        sleep(5)
        
    
        filter_=driver.find_element_by_xpath('/html/body/div[1]/div/main/div[1]/ul/li[2]')
    
        filter_.click()
    
        sleep(9)
        
    except: pass
    
    try:
    
        cell_phones=driver.find_element_by_xpath('/html/body/div[1]/div/main/div[1]/div/div/div[1]/div/div[3]/div/div/div[2]/div/div[1]')
    
        cell_phones.click()
    
        sleep(9)
        
    except: pass
    
    try:
        
        tablet=driver.find_element_by_xpath('/html/body/div[1]/div/main/div[1]/div/div/div[1]/div/div[3]/div/div/div[2]/div/div[2]')

        tablet.click()

        sleep(9)
    
    except: pass
    
    try:
        
        accessories=driver.find_element_by_xpath('/html/body/div[1]/div/main/div[1]/div/div/div[1]/div/div[3]/div/div/div[2]/div/div[3]')

        accessories.click()

        sleep(9)
    
    except: pass
    
    load_all_items()
```

```python
def get_gtin_value():
    
    global gtin_numbers
    
#     wd.support.wait.WebDriverWait.until(10)
    
    x=random.randint(10000000,99999999)
    
    while x in gtin_numbers:
        
        x=random.randint(10000000,99999999)
        
    gtin_numbers.add(x)
    
    return x
```

```python
def stock_and_gtin_value_handler():
    
    global driver
    
    global updated
    
    driver.implicitly_wait(20)
    
#     wd.support.wait.WebDriverWait.until(10)
    
    total_rows=driver.find_elements_by_xpath('//button[@title="More actions"][@class="mc-multi-button mc-multi-button--tertiary"]')
    
    total_rows=total_rows[1:]
    
#     total_rows=total_rows[1250:]
    
#   total_countriesotal_rows=total_rows[1251:]
    
    stock_values=driver.find_elements_by_xpath('//div[@class="mc-table-column mc-table-column--stock mc-table-column--center"]')
    
#     stock_values=stock_values[1250:]
    
    print(len(total_rows),len(stock_values))
    

    for row in range(min(len(total_rows),len(stock_values))):
        
        sv=float(stock_values[row].text)
        
        if sv>0 and sv<=1:

            total_rows[row].click()

            sleep(3)

            edit=total_rows[row].find_elements_by_xpath('//div[@class="mc-multi-button__slot"]')

            edit[0].click()
            
            sleep(5)

            handles=driver.window_handles

            driver.switch_to.window(handles[-1])

            sleep(5)
            
            total_count=total_countries()
            
            x=get_gtin_value()
            
            if total_count>=3:

                gtin=driver.find_element_by_xpath('//input[@name="product-form-gtin"]')

                gtin.clear()

                for i in range(20):

                    gtin.send_keys(Keys.BACKSPACE)

                sleep(1)

                gtin.send_keys(str(x))

                stock=driver.find_element_by_xpath('//input[@name="product-form-stock"]')

                for i in range(10):

                    stock.send_keys(Keys.BACKSPACE)

                stock.send_keys('0')

                category_page_manager()
                
                price_page_handler()

                save=driver.find_element_by_xpath('/html/body/div[1]/div/header/div[2]/div/nav/button[2]')

                save.click()

                sleep(3)
                
                updated+=1
            
            elif total_count<3:
                
                gtin=driver.find_element_by_xpath('//input[@name="product-form-gtin"]')

                gtin.clear()

                for i in range(20):

                    gtin.send_keys(Keys.BACKSPACE)

                sleep(1)

                gtin.send_keys(str(x))

                stock=driver.find_element_by_xpath('//input[@name="product-form-stock"]')

                for i in range(10):

                    stock.send_keys(Keys.BACKSPACE)

                stock.send_keys('0')

                category_page_manager()
                
                variation_page_handler(x)
                
                save=driver.find_element_by_xpath('/html/body/div[1]/div/header/div[2]/div/nav/button[2]')

                save.click()

                sleep(3)
                
                updated+=1

            driver.close()

            driver.switch_to.window(handles[0])
```

```python
def load_all_items():
    
    global driver
    
    driver.implicitly_wait(20)
    
#     wd.support.wait.WebDriverWait.until(20)
    
    page_size=Select(driver.find_element_by_xpath('//select[@name="tab-content__pagesize-selector"]'))
    
    page_size.select_by_visible_text('250')
    
    sleep(5)
    
    total_products=driver.find_element_by_xpath('//div[@class="product-list__product-count"]').text
    
    total_products=total_products.split()
    
    pro=int(total_products[4])
    
    print('Total items found:',pro)
    
    pro=ceil(pro/250)
    
    for i in range(1):

        try:

            load_more=driver.find_element_by_xpath('/html/body/div[1]/div/main/div[3]/button')
            
            load_more.click()
            
            sleep(9)

        except:
            
            print('Loading all items done !')
            
            break
            
    driver.find_element_by_tag_name('body').send_keys(Keys.HOME)
    
    driver.find_element_by_tag_name('body').send_keys(Keys.HOME)
            
#     stock_and_gtin_value_handler()
```

```python
def variation_page_handler(x):
    
    global driver

    variations=driver.find_element_by_xpath('/html/body/div[1]/div/header/div[2]/nav/ul/li[3]')
    variations.click()
    
    sleep(2)
    
    try:
        var=driver.find_elements_by_class_name("variation-table-icon")
    except: pass

    try:
        
        for cur in var:
            cur.click()
            sleep(1)

            gtin_in=driver.find_element_by_xpath('//*[@id="variation-form-gtin"]')

            for i in range(25):
                gtin_in.send_keys(Keys.BACKSPACE)

            gtin_in.send_keys(str(x))

            stock_in=driver.find_element_by_xpath('//*[@id="variation-stock"]')

            for i in range(10):
                stock_in.send_keys(Keys.BACKSPACE)

            stock_in.send_keys('0')

            save_in=driver.find_element_by_xpath('//*[@id="product-form"]/div[3]/div/div[2]/div[4]/div/div/div[3]/div/button[2]')
            save_in.click()
            sleep(1)
            
    except: pass
```
